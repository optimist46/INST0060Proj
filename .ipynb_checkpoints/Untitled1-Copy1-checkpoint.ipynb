```python
#import the full dataset
filepath = 'Data/New1_noQuestionMark.csv'
dataset = import_data()
headers = list(dataset.columns)
```

```python
#create filtering and data split

np.random.seed(2022) #set the seed for the filter, so the random filter is repeatable
train_filter = np.array([True if (i%2 == 0) else False for i in range(len(dataset.iloc[:,0]))])
train_filter = np.random.choice([False, True], len(dataset.iloc[:,0]), p = [0.3, 0.7])
test_filter = ~train_filter

train = dataset.loc[train_filter]
test = dataset.loc[test_filter]

```

```python
y = train.iloc[:,-1:]
X = train.iloc[:,:-1]



n_features = len(train.iloc[0])-1
n_classes = 5
mu = np.zeros((n_classes,n_features))
S_W = np.zeros((n_classes,n_features,n_features))
S_B = np.zeros((n_features,n_features))
i=0
overall_mean = np.mean(X, axis=0).to_numpy()

for cl in np.unique(y):
        y_binary = np.where(y == cl, 1, 0) #'filter' for class
        
        
        
        #calculate the mean values for each class
        mu[i] = np.mean(X[y_binary==1], axis=0)
        
        
        #within class scatter matrix
        S_W[i] = np.cov(X[y_binary==1].T)
        i+=1


```

```python
for cl in np.unique(y):
        # Create a binary class label
        y_binary = np.where(y == cl, 1, 0)

        # Calculate the mean vector for the binary class
        mean_vector = np.mean(X[y_binary==1], axis=0)

        # Calculate the within-class scatter matrix
        S_W = np.cov(X[y_binary==1].T)

        # Calculate the between-class scatter matrix
        overall_mean = np.mean(X, axis=0)
        S_B = mean_vector.dot(mean_vector.T)
```

```python
 # Solve the eigenvalue problem
eigen_values, eigen_vectors = np.linalg.eig(np.linalg.inv(S_W).dot(S_B))

# Make a list of (eigenvalue, eigenvector) tuples
eigen_pairs = np.zeros((5,14))
for i in range(len(eigen_values)):
    eigen_pairs = (np.abs(eigen_values[i]), eigen_vectors[:,i])
eigen_pairs = [(np.abs(eigen_values[i]), eigen_vectors[:,i]) for i in range(len(eigen_values))]

# Sort the (eigenvalue, eigenvector) tuples from high to low
eigen_pairs = sorted(eigen_pairs,  key=lambda k: k[0], reverse=True)

# Select the k eigenvectors with the largest eigenvalues
W = np.zeros((5,13))
for i in range(5):
    W[i] = np.hstack(eigen_pairs[i][1])
    
W = W.T

Predict = X.dot(W)
Predict
```
