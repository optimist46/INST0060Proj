```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
data_t = pd.read_csv('Processed_data.csv') 
data_num = 10000

# classical, hip-hop, pop, r-n-b, edm, rock, acoustic, house, jazz

value = list(range(data_num))
data_t['value'] = value

# Here I assign values to each genre
# Terminal will display some warnings, but it does work
for n in range(data_num):
    if data_t.loc[n][15] == "classical":
        data_t['value'].loc[n] = 1
    elif data_t.loc[n][15] == "hip-hop":
        data_t['value'].loc[n] = 2
    elif data_t.loc[n][15] == "pop":
        data_t['value'].loc[n] = 3
    elif data_t.loc[n][15] == "r-n-b":
        data_t['value'].loc[n] = 4
    elif data_t.loc[n][15] == "edm":
        data_t['value'].loc[n] = 5
    elif data_t.loc[n][15] == "rock":
        data_t['value'].loc[n] = 6
    elif data_t.loc[n][15] == "acoustic":
        data_t['value'].loc[n] = 7
    elif data_t.loc[n][15] == "house":
        data_t['value'].loc[n] = 8
    elif data_t.loc[n][15] == "jazz":
        data_t['value'].loc[n] = 9
    elif data_t.loc[n][15] == "k-pop":
        data_t['value'].loc[n] = 10
print('Change Completed')
```

```python
X_value = []
for n in range(data_num):
    row = list(data_t.loc[n])
    del row[15:17]
    row.append(1)
    X_value.append(row)
print(X_value)
```

Here I directly calculate
<center>$\omega^* = (X^T X)^{-1}X^Ty$
   
The detailed procedure can be obtained in https://www.jianshu.com/p/52583d58e32c (Chinese website).

```python
X = np.mat(X_value)
y = np.mat(list(data_t['value']))
w_b = np.dot((np.dot((np.linalg.inv(np.dot(X.T, X))), X.T)),y.T)
#B = np.linalg.inv(A)
#print(np.dot(A,B))
w_b
```

```python
b = float(w_b[15])
w = np.array(w_b)
w = np.mat(np.delete(w,-1)).T
print(f"w = {w}, b = {b}")
```

Calculate the goodness of fit of predicted values:
<center>$R^2 = 1 - \cfrac{\sum_{i=0}^{n} {(y_i - y_p)^2}}{\sum_{i=0}^{n} {(y_i - \overline{y})^2}}$

```python
# Bset: from sklearn.metrics import r2_score
y_real = list(data_t['value'])
y_mean = np.mean(y_real)

def calc_y_predict(x):
    return float(np.dot(x, w) + b)

RSS = 0
quad_sum = 0
for n in range(data_num):
    x_i = list(data_t.loc[n])
    del x_i[15:17]
    y_p = calc_y_predict(x_i)
    RSS += (y_real[n] - y_p)**2
    quad_sum += (y_real[n] - y_mean)**2
print(RSS, quad_sum)

R2 = 1 - RSS/quad_sum
print(R2)
```

```python
print(f"accuracy = {R2:.2%}")
```
